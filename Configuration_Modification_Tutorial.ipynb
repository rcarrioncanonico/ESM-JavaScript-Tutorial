# Part 1 — Task Overview

Before modifying anything, you need to understand what the task does and what files control it.

## What is the nBackAlpha3 Task?

The **nBackAlpha3** task is a **spatial N-back working memory test**. Participants watch a white square appear in different positions on a 3×3 grid and press `spacebar` when the current position matches the position shown **N trials ago**.

```
┌───────────────────────────────────────────────────────────┐
│                  Task Flow (One Block)                     │
│                                                           │
│  Instructions ──▶ Fixation (1s) ──▶ 30 Trials ──▶ Next   │
│                        │                                  │
│               Each trial (2s total):                      │
│               [Square shown 1s] + [Fixation 1s]           │
│               Participant may press Space anytime          │
│                                                           │
│  Block 1: N-back-1  (match 1 trial ago)                   │
│  Block 2: N-back-2  (match 2 trials ago)                  │
│  Block 3: N-back-3  (match 3 trials ago)                  │
└───────────────────────────────────────────────────────────┘
```

## The 3×3 Grid Layout

The stimulus is a white square that lights up in one of 9 positions:

```
Position numbers:          Coordinates (x, y):
┌─────┬─────┬─────┐        ┌──────────┬──────────┬──────────┐
│  1  │  2  │  3  │        │(-0.2,0.2)│(0.0, 0.2)│(0.2, 0.2)│
├─────┼─────┼─────┤        ├──────────┼──────────┼──────────┤
│  4  │  5  │  6  │        │(-0.2,0.0)│(0.0, 0.0)│(0.2, 0.0)│
├─────┼─────┼─────┤        ├──────────┼──────────┼──────────┤
│  7  │  8  │  9  │        │(-0.2,-0.2)│(0.0,-0.2)│(0.2,-0.2)│
└─────┴─────┴─────┘        └──────────┴──────────┴──────────┘
```

# Part 3 — Configuration Guide

This section documents every parameter you may want to change, where to find it, and what values are safe.

---

## 3.1 Condition Files (`nBackCond1/2/3.xlsx`)

These spreadsheets define the **sequence of trials** for each block. Each row is one trial.

### Column Definitions

| Column | Data Type | What it controls | Valid values |
|--------|-----------|------------------|--------------|
| `square` | Integer | Which grid position lights up (1–9) | `1` through `9` |
| `loc1` | Float | X-coordinate of the square | `-0.2`, `0.0`, `0.2` |
| `loc2` | Float | Y-coordinate of the square | `-0.2`, `0.0`, `0.2` |
| `corrAns` | String or empty | Expected key press | `'space'` or blank/empty |

### Position → Coordinate Lookup

| Position (`square`) | `loc1` (x) | `loc2` (y) |
|---------------------|------------|------------|
| 1 (top-left) | -0.2 | 0.2 |
| 2 (top-center) | 0.0 | 0.2 |
| 3 (top-right) | 0.2 | 0.2 |
| 4 (mid-left) | -0.2 | 0.0 |
| 5 (center) | 0.0 | 0.0 |
| 6 (mid-right) | 0.2 | 0.0 |
| 7 (bottom-left) | -0.2 | -0.2 |
| 8 (bottom-center) | 0.0 | -0.2 |
| 9 (bottom-right) | 0.2 | -0.2 |

### What makes a trial a "target"?

In N-back-1: if trial 5 shows position 6, and trial 4 also showed position 6 → trial 5 is a target → `corrAns = 'space'`

In N-back-2: if trial 7 shows position 3, and trial 5 also showed position 3 → trial 7 is a target.

⚠️ **WARNING:** The `loc1`/`loc2` columns **must match** the `square` column exactly using the lookup table above. If they don't match, the square will appear in the wrong place but the scoring will still use the wrong `corrAns`.

---

## 3.2 Python Script (`nBackAlpha3_lastrun.py`)

This file is generated by PsychoPy Builder and controls the entire task runtime. Key sections:

### Timing Parameters

| Parameter location in script | Default value | What it controls |
|-------------------------------|---------------|------------------|
| `fixation_2.setAutoDraw(True)` start/stop | `0.0` / `1.0` | Inter-block fixation cross duration (1 second) |
| `trialSquare` component duration (trials1) | `0.0` – `1.0` | How long the white square is visible (1 second) |
| `trialFix` component duration (trials1) | `1.0` – `2.0` | How long the fixation shows after the square (1 second) |
| `trialResp` response window | `0.0` – `2.0` | Window in which a keypress is recorded |
| `end_allComponents` draw time | `0.0` – `3.0` | Duration of the end screen (3 seconds) |

### How to find timing values in the script

Search for the pattern: `t >= X and tThisFlipGlobal > (component.tStartRefresh + Y-win.monitorFramePeriod*0.75)`

The `Y` value is the component's **end time** in seconds.

### Trial Count Parameter

| Parameter | Location in script | Default | Controls |
|-----------|--------------------|---------|----------|
| `nReps` for `trials1` | `TrialHandler(... nReps=1)` | `1` | Repetitions of the condition file (30 trials × 1 = 30) |
| `nReps` for `trials2` | same, second block | `1` | N-back-2 repetitions |
| `nReps` for `trials` | same, third block | `1` | N-back-3 repetitions |

💡 **TIP:** Setting `nReps=2` doubles the trials (60 per block). Setting `nReps=0` skips the block entirely — useful for testing only one difficulty level.

### Display Parameters

| Variable name | Default value | What it controls |
|---------------|---------------|------------------|
| `win.color` | `[0, 0, 0]` | Background colour (RGB, range –1 to 1) |
| `trialSquare.fillColor` | `[1, 1, 1]` | Square fill colour (white) |
| `trialSquare.lineColor` | `[1, 1, 1]` | Square border colour |
| `trialSquare.size` | `[0.15, 0.15]` | Square width × height in height units |
| `expInfo['frameRate']` | auto-detected | Monitor refresh rate (do not set manually) |

## 4.1 Create Backups

Run the following cell to create timestamped backups of all modifiable files:

# Part 5 — Step-by-Step Modification Tutorial

This section walks through the mechanics of making changes safely.

---

## 5.1 Editing Timing in the Python Script

### Step 1 — Open the file

📁 Open `nBackAlpha3-master/nBackAlpha3-master/nBackAlpha3_lastrun.py` in VS Code.

### Step 2 — Find the section you want

Use `Ctrl+F` (Find) to search for component names. The script is organised into **routines** (e.g., `trials1`, `trials2`, `trials`).

To find the square display timing for block 1, search for: `trialSquare`

You will find blocks like this:

```python
# *trialSquare* updates
if trialSquare.status == NOT_STARTED and tThisFlip >= 0.0-frameTolerance:
    # keep track of start time/frame for later
    trialSquare.frameNStart = frameN
    trialSquare.tStart = t
    trialSquare.tStartRefresh = tThisFlipGlobal
    win.timeOnFlip(trialSquare, 'tStartRefresh')
    trialSquare.setAutoDraw(True)
if trialSquare.status == STARTED:
    # is it time to stop? (based on global clock, using actual start)
    if tThisFlipGlobal > (trialSquare.tStartRefresh + 1.0-frameTolerance):
        #                                                ^^^
        #                     This is the DURATION in seconds — change this value
        trialSquare.setAutoDraw(False)
```

### Step 3 — Make the change

Identify the number you want to change and edit it. For example, to extend the square display to 1.5 seconds:

**Before:**
```python
if tThisFlipGlobal > (trialSquare.tStartRefresh + 1.0-frameTolerance):
```

**After:**
```python
if tThisFlipGlobal > (trialSquare.tStartRefresh + 1.5-frameTolerance):
```

### Step 4 — Update dependent timings

⚠️ **WARNING:** When you change a component's end time, you must also adjust anything that *starts after it*.

If `trialSquare` ends at `1.5s` instead of `1.0s`, then `trialFix` (the fixation that follows) must start at `1.5s`:

```python
# Before:
if trialFix.status == NOT_STARTED and tThisFlip >= 1.0-frameTolerance:

# After:
if trialFix.status == NOT_STARTED and tThisFlip >= 1.5-frameTolerance:
```

The **total trial duration** must also be updated in the loop-end condition.

### Step 5 — Save the file

`Ctrl+S` in VS Code. Check the title bar — it should no longer show a dot (unsaved indicator).

### Step 6 — Run the task

```bash
conda activate psychopy_env
python nBackAlpha3_lastrun.py
```

### Step 7 — Verify the change

Use a stopwatch on your phone to time the square. If you changed it to 1.5s, you should be able to see clearly that it stays on screen longer.

Alternatively, open the output CSV in `data/` and check the column `trialSquare.stopped - trialSquare.started` — this should be close to your new value.

## 5.3 Editing Number of Trials

There are two ways to change the number of trials per block:

### Method A — Add/remove rows in the condition file (recommended)

Adding rows to `nBackCond1.xlsx` adds trials. Removing rows reduces them.

⚠️ **WARNING:** Maintain the N-back logic. In N-back-1, you cannot have a target on the very first trial because there is no previous trial to match.

### Method B — Change `nReps` in the Python script

Search for `trials1 = data.TrialHandler` in the script:

```python
# Before — runs condition file once (30 trials)
trials1 = data.TrialHandler(nReps=1, method='sequential',
    extraInfo=expInfo, originPath=-1,
    trialList=data.importConditions('nBackCond1.xlsx'),
    seed=None, name='trials1')

# After — runs condition file twice (60 trials)
trials1 = data.TrialHandler(nReps=2, method='sequential',
    ...
```

💡 **TIP:** Set `nReps=0` for a block to skip it entirely. This is useful for testing only the N-back-2 block.

# Part 8 — Verification Utilities

These code cells help you check your configurations programmatically without running the full task.

## 8.1 Check Condition File Consistency

````python
import pandas as pd
import os

# Coordinate lookup: square position → (loc1, loc2)
POSITION_MAP = {
    1: (-0.2,  0.2),
    2: ( 0.0,  0.2),
    3: ( 0.2,  0.2),
    4: (-0.2,  0.0),
    5: ( 0.0,  0.0),
    6: ( 0.2,  0.0),
    7: (-0.2, -0.2),
    8: ( 0.0, -0.2),
    9: ( 0.2, -0.2),
}

TASK_DIR = os.path.join("..", "nBackAlpha3-master", "nBackAlpha3-master")


def check_condition_file(filepath, n_back_level):
    """Validate a condition file for coordinate consistency and N-back target logic."""
    df = pd.read_excel(filepath)
    errors = []
    warnings = []

    print(f"\n{'='*60}")
    print(f"Checking: {os.path.basename(filepath)}")
    print(f"Rows (trials): {len(df)}")
    print(f"Columns: {list(df.columns)}")
    print(f"{'='*60}")

    for i, row in df.iterrows():
        trial_num = i + 1
        sq = row.get('square')
        l1 = row.get('loc1')
        l2 = row.get('loc2')
        corr = row.get('corrAns')

        # Check square value is valid
        if sq not in POSITION_MAP:
            errors.append(f"Trial {trial_num}: 'square'={sq} is not a valid position (must be 1-9)")
            continue

        # Check coordinates match
        expected_l1, expected_l2 = POSITION_MAP[sq]
        if round(float(l1), 4) != expected_l1:
            errors.append(f"Trial {trial_num}: square={sq} → loc1 should be {expected_l1}, got {l1}")
        if round(float(l2), 4) != expected_l2:
            errors.append(f"Trial {trial_num}: square={sq} → loc2 should be {expected_l2}, got {l2}")

        # Check N-back target logic
        if trial_num <= n_back_level:
            if str(corr).strip() == 'space':
                errors.append(f"Trial {trial_num}: corrAns='space' but this is too early to be a target (N-back-{n_back_level})")
        else:
            match_trial = df.iloc[i - n_back_level]
            is_match = (sq == match_trial['square'])
            is_marked_target = str(corr).strip() == 'space'

            if is_match and not is_marked_target:
                warnings.append(f"Trial {trial_num}: square matches {n_back_level} back, but corrAns is NOT 'space'")
            elif not is_match and is_marked_target:
                errors.append(f"Trial {trial_num}: corrAns='space' but square does NOT match {n_back_level} trial(s) ago")

    # Summary
    target_count = sum(1 for _, r in df.iterrows() if str(r.get('corrAns', '')).strip() == 'space')
    target_rate = target_count / len(df) * 100 if len(df) > 0 else 0

    print(f"Targets: {target_count}/{len(df)} ({target_rate:.0f}%)")

    if errors:
        print(f"\n❌ ERRORS ({len(errors)}):")
        for e in errors:
            print(f"  • {e}")
    else:
        print("\n✅ No errors found.")

    if warnings:
        print(f"\n⚠️  WARNINGS ({len(warnings)}):")
        for w in warnings:
            print(f"  • {w}")
    else:
        print("✅ No warnings.")


# Check all three condition files
for level, filename in [(1, 'nBackCond1.xlsx'), (2, 'nBackCond2.xlsx'), (3, 'nBackCond3.xlsx')]:
    path = os.path.join(TASK_DIR, filename)
    if os.path.exists(path):
        check_condition_file(path, n_back_level=level)
    else:
        print(f"⚠️  File not found: {path}")
````

## 8.2 Inspect Timing Values in the Python Script

````python
import re
import os

TASK_DIR = os.path.join("..", "nBackAlpha3-master", "nBackAlpha3-master")
script_path = os.path.join(TASK_DIR, "nBackAlpha3_lastrun.py")

with open(script_path, 'r', encoding='utf-8') as f:
    content = f.read()

# Extract component durations: looks for patterns like
# component.tStartRefresh + X-frameTolerance
pattern = r'(\w+)\.tStartRefresh \+ ([\d.]+)-frameTolerance'
matches = re.findall(pattern, content)

print("Component durations found in script")
print("(these are the end times, relative to component start)")
print("-" * 50)

seen = {}
for component, duration in matches:
    key = f"{component}"
    dur = float(duration)
    if key not in seen:
        seen[key] = []
    seen[key].append(dur)

for comp, durations in sorted(seen.items()):
    unique = sorted(set(durations))
    if len(unique) == 1:
        print(f"  {comp:30s}  {unique[0]:.2f}s  (same across all blocks)")
    else:
        print(f"  {comp:30s}  {unique}  ← different values per block")

# Extract nReps values
print("\n" + "-" * 50)
print("Trial handler nReps values:")
reps_pattern = r"(\w+)\s*=\s*data\.TrialHandler\([^)]*nReps=(\d+)"
reps_matches = re.findall(reps_pattern, content)
for handler, reps in reps_matches:
    status = "✅ enabled" if reps != '0' else "⚠️  DISABLED (nReps=0)"
    print(f"  {handler:20s}  nReps={reps}  {status}")
````

## 8.3 Summarise Output Data

````python
import pandas as pd
import os
import glob

TASK_DIR = os.path.join("..", "nBackAlpha3-master", "nBackAlpha3-master")
data_dir = os.path.join(TASK_DIR, "data")

# Find the most recent CSV file
csv_files = sorted(glob.glob(os.path.join(data_dir, "*.csv")))
if not csv_files:
    print("No CSV files found in data/ — run the task first.")
else:
    latest = csv_files[-1]
    print(f"Most recent output file:\n  {os.path.basename(latest)}\n")

    df = pd.read_csv(latest)

    # Show trial-level summary
    trial_cols = ['trialResp.keys', 'trialResp.corr', 'trialResp.rt', 'square', 'corrAns']
    available = [c for c in trial_cols if c in df.columns]
    trial_data = df[available].dropna(subset=['square'])

    print(f"Total rows (trials): {len(trial_data)}")

    if 'trialResp.corr' in trial_data.columns:
        acc = trial_data['trialResp.corr'].mean()
        print(f"Overall accuracy:    {acc:.1%}")

    if 'trialResp.rt' in trial_data.columns:
        rt = trial_data['trialResp.rt'].dropna()
        if len(rt) > 0:
            print(f"Mean RT (responses): {rt.mean():.3f}s  (SD={rt.std():.3f}s)")

    print("\nFirst 5 trials:")
    print(trial_data[available].head().to_string())
````
