## Pavlovia
- Pavlovia is an open-source, web based plataform which enables scientits to host, run and share online behavioral, psychological, and cognitive experiments. It acts as bridge between the code/tasks written in Psychopy, as it transforms them into web-based (JavaScript) tasks that can be run by participants worldwide, which serves the behavioral science community with secure, cloud-based data storage and management. 

-  Website --> https://pavlovia.org/


## m-Path

- m-Path is the Experience Sampling/EMA (Ecological Momentary Assessment) platform app used for intensive longitudinal methods. It handles the prompting of participants throughout the day (e.g., 3–6 times a day) by sending scheduled notifications to their smartphones. When a participant responds to a prompt, m-Path links them directly to the web-hosted cognitive tasks running on Pavlovia. This makes m-Path the entry point for participants in the EMA data collection pipeline, connecting the sampling schedule to the actual task execution in the browser.

- Website --> https://manual.m-path.io/

## PsychoJS

- PsychoJS is the JavaScript library that allows PsychoPy experiments to be translated into a format that runs directly in a web browser. It serves as the bridge between the Python-based task code written in PsychoPy and the smartphone or browser environment used by participants in the EMA setting. When a PsychoPy experiment is uploaded to Pavlovia, the platform uses PsychoJS to execute it as a web application — meaning any participant can complete the task on their smartphone without installing any software. This translation layer is what makes it possible to deploy Python-based cognitive tasks remotely across the Pavlovia platform.



## Local Lab Infrastructure

Cognitive tasks are also administered locally on research computers in the Tilburg University Lifespan Lab. In this setting, tasks run natively in PsychoPy on a controlled desktop machine rather than through a web browser. This local **Lab** setting is distinct from the remote **EMA** setting: while EMA relies on m-Path, Pavlovia, and PsychoJS to deliver tasks to participants' smartphones in daily life, the Lab setting provides a controlled environment for ground-truth validation — ensuring that the web-based and lab-based versions of each task produce comparable data.

## PsychoPy

Psychopy is a free, open-source software package written in Python, designed for creating psychology, neuroscience, and psychophysics. It is useful for presenting stimuli and collect data using both the GUI-based builder window and the advanced "Coder" scripting which is what we will focus on.

It is flexibile meaning that it allows for large complex experimental designs, as well as having online compatibility with Pavlovia which turns the experiments to JavaScript to run them directly in the web browser. It is also convinient as it runs in Windoes, Mac and Linux.

What is important to know about psychopy is that in order for experiments to run in it, it has to be compatible with other versions of python depending on the experiment. The gold-standard for psychopy coding is python version 3.10 as of 2026. 

Thus is important that the kernel is set to the psyochopy enviroment meanwhile learning about psychopy during this tutorial.

You should download and install the standalone version of PsychoPy in your computer.
- https://www.psychopy.org/download.html
