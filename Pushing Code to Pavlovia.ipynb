Steps to Push Code to Pavlovia from VS Code
Create a New Project on Pavlovia:
Log in to pavlovia.org and create a new, empty repository.
Clone the Repository Locally:
Copy the Git URL of your new repository.
Open the terminal in VS Code and clone it to your local machine:
git clone <repository_url>
Set Up Your Experiment Files:
Move your JavaScript/HTML/Psychopy files into the cloned folder.
If using jsPsych, ensure the index.html and supporting files are present.
Note: If using PsychoPy, you must have the necessary HTML and JS files in the repository.
Push Changes Using VS Code Git:
Open the Source Control panel (Ctrl+Shift+G).
Stage your changes (click '+').
Commit your changes with a message.
Click '...' and select "Push" to send your changes to the GitLab repository.

Important Considerations
Gitlab Credentials: You will likely need to log in with your Pavlovia username and password/token when pushing for the first time.
Sync Button Alternative: If the standard "Sync" button in PsychoPy fails, using Git commands directly in a terminal is a valid workaround to push changes.
Structure: Ensure your index.html is at the root directory of the repository.
