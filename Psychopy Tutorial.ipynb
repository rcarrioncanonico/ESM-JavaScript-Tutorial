# How to Code In VS Code for Psychopy

In coding, an environment is like a dedicated kitchen where you keep specific ingredients for a recipe.

A PsychoPy "recipe" (your experiment code) only works if it is run in the kitchen it was made for. In our case, the `psychopy_env` environment is required because it contains the right tools and ingredients for the code.

To work efficiently in this kitchen, we need to set up the PsychoPy environment for VS Code. VS Code provides helpful tools like GitHub Copilot, which makes editing and creating code easier. There are different ways to do this, but this tutorial uses Miniconda so you can manage multiple environments (for example, one for PsychoPy and another for EEG analysis).

Install Miniconda here: https://www.anaconda.com/docs/getting-started/miniconda/main

Download the PsychoPy environment file here: https://github.com/rcarrioncanonico/Momentary-Assesment-Tools/blob/main/psychopy_env.yml

After installing Miniconda, open the "Anaconda Prompt" application to create the PsychoPy environment. You can find it by pressing the Windows key and typing "Anaconda Prompt". Then run:

```
conda env create -f psychopy_env.yml
```

### Note

If `psychopy_env.yml` is not found, you are in the wrong directory. Either move the file to the directory shown in Anaconda Prompt, or change directories. For example, to switch to Downloads:

```
cd ~\Downloads
```

As mentioned in the Introduction Tutorial, if you cannot run commands in Anaconda Prompt, it is likely due to Windows security policy. Open PowerShell as Administrator and run:

```
Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser
```

# Towards VS Code

You should now have an environment named `psychopy_env`. Press the Windoes key and look for `miniconda3` > `envs` and confirm a `psychopy_env` folder exists.

For coding tasks in VS Code, set the interpreter and terminal to that same environment:

1) Interpreter: click the Python interpreter selector in the bottom-right status bar (next to the AI icon) and choose `psychopy_env`.
2) Terminal: open Terminal > New Terminal. It should activate `psychopy_env` by itself. If it does not, run:

```
conda activate psychopy_env
```

# Troubleshooting: The 'Front Door' vs. the 'Chef' Logic

When working in your terminal, it's important to understand the difference between two environments:

**PowerShell Terminal (the 'Front Door')**  
This is the main entry point to your computer. It is where you run system commands and file operations (like `conda activate psychopy_env`). The prompt looks like:  
```
PS C:\Users\...>
```

**Python REPL (the 'Chef')**  
Once you type `python` (which is a command to run your python file/ task) in the PowerShell terminal, you enter the 'Chef' — an interactive Python environment where you can run Python code directly. The prompt looks like:  
```
>>> 
```

**Important:** You cannot run script commands like `python file.py` while you are in the 'Chef' (Python REPL). If you see the `>>>` prompt, you are inside the 'Chef', not the 'Front Door'. You must type the following to exit and return to PowerShell:

```
exit()
```

Once you see the PowerShell prompt again, you can run your Python scripts and conda commands.

# Opening Your Project Folder

For experiments that use external assets (images, sounds, CSV files, etc.), the **Working Directory** must be set correctly. To do this in VS Code:

Open your experiment folder directly:  
1. Go to File > Open Folder
2. Select the folder containing your experiment files
3. This ensures that all file paths in your Python code are relative to this folder

**Why is this important?** When your code references a file like `images/stimulus.png`, Python needs to know where to look. By opening the correct folder, the working directory is automatically set, and all relative paths will work correctly.

**⚠️ Warning:** If you move files out of their experiment folder, or if the folder structure changes, your code will fail with a `FileNotFoundError`. Always keep your files organized and avoid moving them outside the experiment folder.

# Advanced 'PsychoPy Sanity Check' Exercise

Before running your full experiment, always test that PsychoPy is properly installed and that your graphics environment is working. This code will:

1. Initialize a PsychoPy window
2. Display 'PsychoPy is Working!'
3. Wait for 2 seconds
4. Close the window

If you see a black window with text, your kitchen is ready. If you see an error, the installation needs troubleshooting (you might need to run this code twice to ensure it works properly).

````python
from psychopy import visual, core
import time

# Create a PsychoPy window (black background)
win = visual.Window([800, 600], color='black', monitor='testMonitor')

# Create a text stimulus
text = visual.TextStim(win, text='PsychoPy is Working!', color='white', height=0.1)

# Draw the text and flip the window to display it
text.draw()
win.flip()

# Wait for 2 seconds
time.sleep(2)

# Close the window
win.close()
core.quit()

print('PsychoPy sanity check complete!')
````

# Environment Blueprinting & Reproducibility

**The Kitchen Blueprint**

The `psychopy_env.yml` file is the 'blueprint' of your kitchen. It lists every ingredient and tool needed for PsychoPy to work. This is crucial for reproducibility.

**For Team Collaboration and Version Control:**

Keep `psychopy_env.yml` in the root of your GitHub repository. This way, any team member (or your future self) can recreate the exact environment by running:

```
conda env create -f psychopy_env.yml
```

This ensures that everyone uses the same versions of PsychoPy, Python, and all dependencies. No more 'it works on my machine' problems!

**Best Practice:** Commit `psychopy_env.yml` to version control alongside your code. When the lab updates PsychoPy or adds new tools, update the `.yml` file and share it with collaborators.
