# Lesson 3 — Functions (JavaScript)

Functions are the main "building blocks" of a program. They allow the code to be called many times without repetition. You have already seen examples of built-in functions like `console.log()` and `Number()`. Now, we will learn how to create our own.

## Learning objectives

By the end of this lesson you can:
- Declare and call a function.
- Pass data into functions using parameters.
- Return data from functions.
- Understand the difference between local and outer variables (scope).

## 1) Function Declaration

To create a function we can use a function declaration.

It looks like this:
```javascript
function name(parameters) {
  // code body
}
```

````javascript
function showMessage() {
  console.log("Hello everyone! This is my first function.");
}

// To execute the function, you call it by its name followed by parentheses:
showMessage();
showMessage(); // We can call it as many times as we want!
````

## 2) Local and Outer Variables

A variable declared **inside** a function is only visible inside that function (Local variable).
A function can also access variables declared **outside** of it.

````javascript
let userName = "Alice"; // Outer variable

function greet() {
  let localMessage = "Hello, " + userName;
  console.log(localMessage);
}

greet();
// console.log(localMessage); // This would cause an error because localMessage is not visible outside the function.
````

## 3) Parameters and Default Values

We can pass arbitrary data to functions using parameters (also called arguments).
If a parameter is not provided when the function is called, its value becomes `undefined`. We can provide a "default" value using `= ...`.

````javascript
function showGreeting(from, text = "no text given") {
  console.log(from + ": " + text);
}

showGreeting("Ann", "Hello!"); // Ann: Hello!
showGreeting("Ann"); // Ann: no text given
````

## 4) Returning a Value

A function can return a value back into the calling code as the result.
When execution reaches `return`, the function stops, and the value is returned to the calling code.

````javascript
function sum(a, b) {
  return a + b;
}

let result = sum(5, 10);
console.log("The sum is: " + result);
````

# Solutions

## Exercise 1

````javascript
function min(a, b) {
  if (a < b) {
    return a;
  } else {
    return b;
  }
}

// A shorter way using Math.min:
// function min(a, b) { return Math.min(a, b); }

console.log(min(2, 5));
````

## Exercise 2

````javascript
function isValidTrial(rt) {
  if (rt > 100 && rt < 1000) {
    return true;
  }
  return false;
}

// A much shorter way:
// function isValidTrial(rt) { return rt > 100 && rt < 1000; }

console.log("Try 50:", isValidTrial(50));
console.log("Try 500:", isValidTrial(500));
````

# Explanations

## Exercise 1:
We can use an `if...else` statement inside our function to compare `a` and `b`. The `return` statement immediately passes the result back and exits the function. Built-in functions like `Math.min()` can also do this for us, but writing it manually helps understand the underlying logic.

## Exercise 2:
We use the logical AND operator (`&&`) to check multiple conditions simultaneously: `rt` must be greater than 100 AND `rt` must be less than 1000. If both are true, it returns `true`. If execution reaches past the `if` block, it returns `false`.
