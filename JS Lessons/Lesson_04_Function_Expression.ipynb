# Lesson 4 — Function Expressions (JavaScript)

In JavaScript, a function is not just a magical language structure—it is a special kind of value. This means you can treat functions like any other value: assign them to variables, pass them as arguments, or return them from other functions.

## Learning objectives

By the end of this lesson you can:
- Create a Function Expression.
- Understand that functions are values.
- Pass functions as arguments to other functions (Callback functions).
- Explain the difference between a Function Declaration and a Function Expression (Hoisting).

## 1) Function Expressions

Previously, we learned about Function Declarations (`function sayHi() {...}`). There is another syntax for creating a function, called a **Function Expression**.

It allows us to create a new function in the middle of any expression, such as assigning it to a variable.

````javascript
let sayHi = function() {
  console.log("Hello!");
}; // Note the semicolon at the end of the assignment

sayHi();

// Because it's just a value, we can copy it to another variable:
let funcCopy = sayHi;
funcCopy(); // Also says "Hello!"
````

## 2) Callback Functions

Since functions are just values, we can pass them as arguments to *other* functions.
A function that is passed as an argument and called later is known as a **callback function** (or just callback).

````javascript
function runTask(taskName, onComplete) {
  console.log("Running task: " + taskName);
  // "Call back" the provided function when done
  onComplete();
}

function finishMessage() {
  console.log("Task has been completed successfully!");
}

// We pass the finishMessage function WITHOUT parentheses to pass the function itself, not its result.
runTask("Download Data", finishMessage);
````

Alternatively, we can write the function directly inside the argument list. This is called an **anonymous function** because it doesn't have a name.

````javascript
runTask("Update Screen", function() {
  console.log("Screen updated directly via anonymous callback!");
});
````

## 3) Function Declaration vs Function Expression

The key difference between the two is **when** they are created by the JavaScript engine.

- **Function Declarations** are processed *before* the code block is executed. They are "hoisted" to the top, meaning you can call them earlier in the code than where they get defined.
- **Function Expressions** are created when the execution reaches them. They cannot be used before they are assigned.

````javascript
// Function Declaration (Hoisted)
greetDeclaration(); // This WORKS!

function greetDeclaration() {
  console.log("Hi from Declaration");
}

// Function Expression (Not Hoisted)
// greetExpression(); // This would throw an error!

let greetExpression = function() {
  console.log("Hi from Expression");
};

greetExpression(); // This works, because it's called after creation
````
