# Solutions

## Exercise 1

````typescript
let calculator = {
    read() {
        this.a = +prompt("a?", 0)
        this.b = +prompt("b?", 0)
    },
    sum() {
        return this.a + this.b
    },

    mul() {
        return this.a * this.b
    },
        
};

calculator.read();
console.log( "Sum: ", calculator.sum() );
console.log( "Product: ", calculator.mul() );
````

# Explanations:


## Exercise 1:


When creating and object method, we first create the object, in this case let calculator = {}, and then we add methods to it. Each method is a function that can access the properties of the object using `this.`. In the exercise above, we created three methods: `read()`, `sum()`, and `mul()`. The `read()` method prompts the user for input and saves it as properties a and b. The `sum()` method returns the sum of a and b, while the `mul()` method returns their product. You might notice that we use `this.a` and `this.b` when defining the methods of the calculator object. This is because `this`refers to the properties of the object itself, allowing us to access and manipulate the values stored in a and b across different methods. We firstly defined a and b as properties of the object by using `this.a` and `this.b` in the read() method, and then we can use those properties in the `sum()` and `mul()` methods to perform calculations.

Why `+prompt()`and not just `prompt()`?. Because `prompt()` returns a string, and the + operator converts it to a number. If you omit the +, the values will be treated as strings, and `sum()` will concatenate them instead of adding them numerically. For example, if you enter 2 and 3, sum() would return "23" instead of 5 without the +.

`Prompt()` function can take two arguments: prompt(text, default). The first one is the text to show the user, and the second one is the default value. In this case, we use 0 as the default value to ensure that if the user doesn't enter anything, it will be treated as 0 in calculations.

In JavaScript, when you are inside a method (like sum()), the variables a and b are not automatically recognized as belonging to the object.

If you just write return a + b;, JavaScript thinks you are looking for local variables named a and b defined directly inside that specific function. Since they aren't defined inside sum().


The keyword this is required because it tells JavaScript to look at the object itself.

## Exercise 2:

`What is ++ in this.step++?`

The ++ operator is a shorthand for incrementing a variable by 1. In this case, this.step++ is equivalent to this.step = this.step + 1. It increases the value of the step property by 1 each time the up() method is called. So when you call up() on the ladder object, it will move you up one step by incrementing the step property. Similarly, in the down() method, we use this.step-- to decrement the step property by 1, allowing you to move down the ladder.

` Why return this;` when defining the up() and down() methods?

Returning this from the `up()` and `down()` methods allows for `method chaining`. When a method returns this, it returns the object itself, which means you can call another method on the same object immediately after. For example, you can do ladder.up().up().down() to move up two steps and then down one step in a single line of code. This makes the code more concise and allows for a more fluent interface when working with the ladder object.

We didnt not have to return `this` (the whole object) during the calculator exercise because we were not chaining methods together. In the calculator example, each method (read(), sum(), mul()) is called separately and does not need to return the object itself for further method calls. However, in the ladder example, we want to allow for chaining multiple method calls together, which is why we return this from the up() and down() methods. If we were to return this; inside sum() when calling calculator.sum(), it would give you back the entire calculator object instead of the result of the sum, which is not what we want in that context.

In the ladder object `up()` and `down()` are just actions that change the internal state of `this.step`. When you take a step up you don´t need a specific value back, you just want to change the state of the ladder, specifically the property `step`. The purpose of returning the whole object (`this`) is to allow for method chaining, which is a common pattern in JavaScript that allows you to call multiple methods on the same object in a single line of code. In the case of the ladder, you might want to call up() multiple times in a row, and returning this allows you to do that seamlessly. In contrast, in the calculator example, each method is designed to return a specific value (the sum or product), so returning the whole object would not be appropriate in that context.


# Related Information:

## What is an Object Literal?

Both let calculator = { ... } and let ladder = { ... } in your exercises are examples of Object Literals. An object literal is simply a comma-separated list of name-value pairs wrapped in curly braces {}. It is the easiest and most common way to create an object in JavaScript right on the spot.

When you literally type out the curly braces and write the properties and methods inside them, you are using the object literal syntax.

syntax:
```javascript
let ladder = {
    step: 0,
    up() {
        this.step++;
        return this;
    }
};
```
### Why does it matter?

There are other ways to create objects in JavaScript, such as using constructor functions or the Object.create() method. However, object literals are often more concise and easier to read for simple objects that don't require complex inheritance or behavior. They allow you to define an object with its properties and methods in a clear and straightforward way, making them a popular choice for creating objects in JavaScript.

1. The new Object() syntax is more verbose and less efficient than using an object literal. It creates an empty object and then adds properties to it, which can be less performant than defining the object with its properties directly in an object literal. 
```javascript
let ladder = new Object();
ladder.step = 0;
ladder.up = function() {
    this.step++;
    return this;
};
```
2. Constructor Functions or Classes: (Used when you want to make many ladders, rather than just one). 
```javascript
class Ladder {
    constructor() {
        this.step = 0;
    }
    up() {
        this.step++;
        return this;
    }
}
let myLadder = new Ladder();
```
