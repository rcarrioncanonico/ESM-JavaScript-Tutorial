# Lesson 05 - Arrays, Objects, and Stimulus Lists

In experimental design, you almost always need a **Stimulus List**. This acts as the master blueprint for your task block. 

A Stimulus List is simply an **Array** containing multiple **Objects**. 
- An **Object** (or dictionary) groups associated variables for a *single trial* together (e.g., `{condition: 'incongruent', text: 'BLUE', color: 'red', correctKey: 'r'}`).
- An **Array** (or list) holds the ordered sequence of these trial objects.

This lesson bridges these two fundamental JavaScript data structures. Understanding how to navigate arrays of objects is critical for feeding your experimental parameters into the UI sequentially, and eventually storing nested structures holding the participant's tracking metrics (Reaction Time, Accuracy, etc.) to export as JSON data at the end of the psychological task.

## Learning objectives

By the end of this lesson you can:
- Group related trial features (condition, target, correct response) into JavaScript objects.
- Store a full task schedule as an aggregate list (an array of objects).
- Safely read object fields during a live experiment to map conditions to task stimuli.
- Understand how nested data structures help organize trial results (accuracy, RT, side) for eventual data export.

## Apply to `student_task.js`

Goal: cycle through a trial array and show current stimulus text.

Required elements:
- array of trial objects,
- `trialIndex`,
- read `trials[trialIndex].stimulus`.

### Starter snippet

```javascript
const trials = [
  { stimulus: "LEFT", correctKey: "ArrowLeft" },
  { stimulus: "RIGHT", correctKey: "ArrowRight" },
  { stimulus: "LEFT", correctKey: "ArrowLeft" }
];

let trialIndex = 0;
const label = this.add.text(400, 220, "Press SPACE", {
  fontSize: "30px",
  color: "#ffffff"
}).setOrigin(0.5);

this.input.keyboard.on("keydown-SPACE", () => {
  if (trialIndex < trials.length) {
    label.setText("Stimulus: " + trials[trialIndex].stimulus);
    trialIndex = trialIndex + 1;
  } else {
    label.setText("All trials shown");
  }
});
```
