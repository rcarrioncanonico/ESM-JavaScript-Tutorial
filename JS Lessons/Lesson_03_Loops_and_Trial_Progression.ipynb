# Lesson 03 - Loops and Trial Progression (JavaScript)

In cognitive psychology paradigms, participants often complete dozens or hundreds of trials. Manually writing out the logic or configuration for every single trial is obviously impossible. This lesson teaches you how to harness **loops** in JavaScript to automate repetitive processes. 

Loops are essential for ESM tasks to:
1. **Pre-generate trial sequences:** Instantly create 50 or 100 trial configurations (with randomized conditions or jittered Inter-Stimulus Intervals) before the task even starts.
2. **Iterate over data:** Loop through a participant's responses at the end of a block to calculate their average accuracy.
3. **Control trial flow:** Advance an `index` variable step-by-step through a pre-calculated sequence, updating the UI safely during runtime.

## Learning objectives

By the end of this lesson you can:
- Use `for` loops to repeatedly handle task actions or generate sequences.
- Create an array of trial components automatically to avoid duplicating code.
- Understand the difference between generating a task sequence upfront using loops vs traversing it sequentially during runtime.
- Update UI labels iteratively to reflect trial progression.

## Apply to `student_task.js`

Goal: Show trial progress and update on `SPACE`.

Required elements:
- `for` loop to prebuild trial labels,
- `trialIndex` state,
- label updates with `.setText(...)`.

### Starter snippet

```javascript
let trialIndex = 0;
const maxTrials = 5;
const trialLabels = [];

for (let i = 1; i <= maxTrials; i = i + 1) {
  trialLabels.push("Trial " + i + " / " + maxTrials);
}

const statusLabel = this.add.text(400, 220, "Press SPACE to start", {
  fontSize: "26px",
  color: "#ffffff"
}).setOrigin(0.5);

this.input.keyboard.on("keydown-SPACE", () => {
  if (trialIndex < maxTrials) {
    statusLabel.setText(trialLabels[trialIndex]);
    trialIndex = trialIndex + 1;
  } else {
    statusLabel.setText("Task complete");
  }
});
```
