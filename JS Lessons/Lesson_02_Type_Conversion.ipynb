# Lesson 2 — Type Conversions (JavaScript)

This lesson covers how JavaScript converts values between different data types. Most of the time, operators and functions automatically convert values given to them to the right type, but sometimes we need to do it explicitly to avoid bugs.

## Learning objectives

By the end of this lesson you can:
- Convert values to Strings explicitly and implicitly.
- Convert values to Numbers properly and identify when `NaN` is produced.
- Understand the rules of Boolean conversion (what is truthy vs falsy).
- Avoid common bugs when dealing with `undefined`, `null`, and `"0"`.

## 1) String Conversion

String conversion happens when we need the string form of a value. For example, when you output something or concatenate values.
We can also call the `String(value)` function to convert a value to a string explicitly.

```javascript
let value = true;
console.log(typeof value); // "boolean"

value = String(value); // now value is a string "true"
console.log(typeof value); // "string"
```

## 2) Numeric Conversion

Numeric conversion happens automatically in mathematical functions and expressions (like division `/` or multiplication `*`).

We can use the `Number(value)` function to explicitly convert a value to a number. This is usually required when we read a value from a string-based source (like user input) but expect a number.

```javascript
let str = "123";
console.log(typeof str); // "string"

let num = Number(str); // becomes a number 123
console.log(typeof num); // "number"
```

If the string is not a valid number, the result of such a conversion is `NaN` (Not a Number).

Rules for numeric conversion:
- `undefined` becomes `NaN`
- `null` becomes `0`
- `true` / `false` becomes `1` / `0`
- **string**: Whitespaces from the start and end are removed. If the remaining string is empty, the result is `0`. Otherwise, the number is read from the string. An error gives `NaN`.

```javascript
console.log(Number("   123   ")); // 123
console.log(Number("123z"));      // NaN (error reading a number at "z")
console.log(Number(true));        // 1
console.log(Number(false));       // 0
console.log(Number(null));        // 0
console.log(Number(undefined));   // NaN
```

## 3) Boolean Conversion

Boolean conversion occurs in logical operations or can be performed explicitly with `Boolean(value)`.

The conversion rules are simple:
- Values that are intuitively "empty", like `0`, an empty string `""`, `null`, `undefined`, and `NaN`, become `false` (also known as falsy values).
- Other values become `true` (truthy values).

```javascript
console.log(Boolean(1)); // true
console.log(Boolean(0)); // false
console.log(Boolean("hello")); // true
console.log(Boolean("")); // false

// Important special cases:
console.log(Boolean("0")); // true, because non-empty strings are always true in JavaScript
console.log(Boolean(" ")); // spaces, also true (any non-empty string is true)
```

# Explanations

## Exercise 1:
To explicitly convert a value to a string, you pass it into the `String(value)` function. This is highly useful when we want to display numbers as text or log them in specific string-based formats, guaranteeing the variable has a string type.

## Exercise 2:
The `+` operator in JavaScript performs string concatenation if any of its operands is a string (`"250" + "500" === "250500"`). By wrapping the values in the `Number()` function, you explicitly convert both strings into proper numbers before adding them. A common shorthand for `Number(value)` is placing a `+` directly before the variable, like `+baseline`.

## Exercise 3:
Values like `null`, `undefined`, and `""` (an empty string) are considered "empty" or "falsy" in JavaScript, so they evaluate to `false`. Conversely, a string containing only a space (`" "`) or the string `"0"` are **non-empty strings**, so they evaluate to `true`. This trips up many beginners since in some other languages `"0"` might evaluate to false!
