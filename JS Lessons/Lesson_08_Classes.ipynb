# Lesson 8 — Classes (JavaScript)

In real life, we often need to create many objects of the same kind, like users, game characters, or trials in an experiment. 

While we can create single objects using `{...}`, JavaScript provides a more advanced `class` construct. A class acts as a blueprint for creating multiple objects that share the same properties and methods.

## Learning objectives

By the end of this lesson you can:
- Declare a class using the `class` keyword.
- Understand the role of the `constructor()` method.
- Add methods to a class.
- Create new object instances from a class using the `new` keyword.

## 1) The "class" syntax

The basic syntax is:

```javascript
class MyClass {
  // constructor method is called automatically when "new" is used
  constructor() { ... }
  
  // other methods
  method1() { ... }
  method2() { ... }
}
```

Then we use `new MyClass()` to create a new object with all the listed methods.

````javascript
class User {
  constructor(name) {
    this.name = name;
  }

  sayHi() {
    console.log("Hello, " + this.name);
  }
}

// Usage:
let user = new User("John");
user.sayHi(); // Hello, John
````

What exactly does `new User("John")` do?
1. A new empty object is created.
2. The `constructor` runs with the given argument (`"John"`) and assigns it to `this.name`.
3. The object gains access to the methods defined in the class (like `sayHi`).

`How does constructor() work?`
The `constructor()` method is a special function that runs automatically when a new instance of the class is created. It initializes the object's properties. Inside the parameters of `constructor()`, we can define any properties we want our objects to have. But this is only for dynamically assigned properties. If we want to assign a property that is the same for all objects, we can do that outside of the constructor. For example, if we want all users to have a `role` property that is set to `"user"`, we can define it directly in the class body.
```javascript
class User {
  role = "user"; // This property will be the same for all instances

  constructor(name) {
    this.name = name; // This property will be unique for each instance
  }
}
```In this example, every instance of `User` will have a `role` property set to `"user"`, while the `name` property will be unique to each instance based on the argument passed to the constructor.

**Class Fields vs. Constructor Properties**

You might wonder why we sometimes initialize properties inside the constructor (like `this.name = name`) and sometimes outside of it (like `role = "user"`).

- **Inside the constructor (Traditional):** Best for properties that depend on arguments passed when creating the object (dynamically assigned).
- **Outside the constructor (Class Fields):** Best for properties that always start with a fixed default value (like `score = 0` or `role = "user"`). This modern approach is cleaner and immediately shows what default properties every new object gets.

## 2) No commas between methods

A common pitfall for beginners is putting a comma between class methods. Do not put commas here, it will result in a syntax error. Classes are not object literals!

````javascript
class BadExample {
  method1() {}
  //, <--- DO NOT ADD COMMAS BETWEEN METHODS
  method2() {}
}
````

# Solutions

## Exercise 1

````javascript
//Traditional Method
class Player {
  constructor(name) {
    this.name = name;
    this.score = 0;
  }

  addScore(points) {
    this.score += points;
  }

  showScore() {
    console.log("Player " + this.name + " has " + this.score + " points.");
  }
}

let player1 = new Player("Alice");
player1.addScore(10);
player1.addScore(5);
player1.showScore();
````

````javascript
//Class field method
class Player {
  score = 0; // Using a class field for the default score

  constructor(name) {
    this.name = name;
  }

  addScore(points) {
    this.score += points;
  }

  showScore() {
    console.log("Player " + this.name + " has " + this.score + " points.");
  }
}

let player1 = new Player("Alice");
player1.addScore(10);
player1.addScore(5);
player1.showScore();
````

# Explanations

## Exercise 1:
The `constructor(name)` method is automatically triggered when we type `new Player("Alice")`. Inside the constructor, `this` refers to the newly created player object, allowing us to initialize `this.name = "Alice"` and set `this.score = 0`. 

Once the object is created, we can call `addScore(10)`, which uses `this.score += points` to update the active object's score. Since we didn't add commas between methods in the `class` block, everything compiles nicely into an organized blueprint.
