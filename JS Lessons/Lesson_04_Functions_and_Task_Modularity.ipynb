# Lesson 04 - Functions and Task Modularity (JavaScript)

This lesson shows how to move repeated logic into functions so your task code is cleaner and easier to debug.

## Learning objectives

By the end of this lesson you can:
- Create custom functions to bundle repetitive code, making the task logic cleaner.
- Pass variables into functions (parameters) to handle dynamic trial states properly.
- Return calculated values from functions back to the main game loop.
- Separate logic into modular blocks (e.g., drawing UI vs tracking trial outcomes).
- Grasp how callbacks and arrow functions execute during task events.

````typescript
// Warm-up: function for trial text
function formatTrialText(trialIndex, maxTrials) {
  return "Trial " + trialIndex + " / " + maxTrials;
}

formatTrialText(2, 5)
````

## Apply to `student_task.js`

Goal: use functions for status updates and feedback color.

Required elements:
- one function that returns status text,
- one function that applies text + color to a label.

### Starter snippet

```javascript
let trialIndex = 0;
const maxTrials = 4;

const label = this.add.text(400, 220, "", {
  fontSize: "26px",
  color: "#ffffff"
}).setOrigin(0.5);

function getStatusText(index, max) {
  if (index === 0) return "Waiting to start";
  if (index < max) return "Trial running";
  return "Task complete";
}

function applyStatus(labelObj, text) {
  labelObj.setText(text);
  if (text === "Task complete") {
    labelObj.setColor("#88ff88");
  } else {
    labelObj.setColor("#ffffff");
  }
}

applyStatus(label, getStatusText(trialIndex, maxTrials));

this.input.keyboard.on("keydown-SPACE", () => {
  if (trialIndex < maxTrials) trialIndex = trialIndex + 1;
  applyStatus(label, getStatusText(trialIndex, maxTrials));
});
```
