# Lesson 06 - Keyboard Events and Reaction Time

Accurate behavioral measurement relies heavily on high-fidelity user input mechanisms. In phone or web-based cognitive batteries, you must accurately capture *what* the user pressed and *exactly when* they pressed it.

This lesson introduces how browser interfaces capture hardware interrupts (keyboard events) and calculate time differentials. 

We will cover:
- **Event Listeners:** Telling Phaser to actively listen for specific (or general) key inputs.
- **Precision Timing:** Moving past standard `Date.now()` constraints to utilize `performance.now()`, which grants the sub-millisecond precision required to differentiate cognitive load.
- **Accuracy Logic:** Comparing the key the user *actually* pressed with the `correctKey` parameter configured in the target trial, and dynamically responding with visual feedback.

## Learning objectives

By the end of this lesson you can:
- Register keyboard event listeners in Phaser to capture user key presses cleanly.
- Evaluate response accuracy continuously by comparing logged keys against anticipated targets.
- Access high-precision timestamps (`performance.now()`) to measure reaction time precisely.
- Differentiate stimulus timing windows inside game events (trial onset, response deadline, trial offset).

## Apply to `student_task.js`

Goal: show a target letter, then report correctness + RT after key press.

### Starter snippet

```javascript
const expectedKey = "A";
const startTime = Date.now();

const prompt = this.add.text(400, 180, "Press A", {
  fontSize: "34px",
  color: "#ffffff"
}).setOrigin(0.5);

const feedback = this.add.text(400, 260, "", {
  fontSize: "26px",
  color: "#aaddff"
}).setOrigin(0.5);

this.input.keyboard.on("keydown", (event) => {
  const endTime = Date.now();
  const rtMs = endTime - startTime;
  const isCorrect = event.key.toUpperCase() === expectedKey;

  if (isCorrect) {
    feedback.setText("Correct | RT: " + rtMs + " ms");
    feedback.setColor("#88ff88");
  } else {
    feedback.setText("Incorrect | RT: " + rtMs + " ms");
    feedback.setColor("#ff9999");
  }
});
```
