# Lesson 9 — JSON (JavaScript)

Let's say we have a complex object, and we'd like to convert it into a string to send it over a network or save it to a database (like saving a player's trial data). 

JSON (JavaScript Object Notation) is a general format to represent values and objects. It's widely used for data exchange.

## Learning objectives

By the end of this lesson you can:
- Understand what JSON is and why it's used.
- Convert JavaScript objects into JSON strings using `JSON.stringify`.
- Convert JSON strings back into JavaScript objects using `JSON.parse`.

## 1) JSON.stringify

The `JSON.stringify` method takes an object (or an array, or even a primitive) and converts it into a string.

```javascript
let student = {
  name: "John",
  age: 30,
  isAdmin: false,
  courses: ["html", "css", "js"],
  spouse: null
};

let json = JSON.stringify(student);

console.log(typeof json); // string!
console.log(json);
// Output:
// {"name":"John","age":30,"isAdmin":false,"courses":["html","css","js"],"spouse":null}
```

Please note that a JSON-encoded object has several important differences from an object literal:
- Strings use double quotes. No single quotes or backticks. So `'John'` becomes `"John"`.
- Object property names are also double-quoted. `age:30` becomes `"age":30`.

## 2) JSON.parse

To decode a JSON-string, we need another method named `JSON.parse`.

```javascript
// A stringified array
let numbers = "[0, 1, 2, 3]";
let parsedNumbers = JSON.parse(numbers);
console.log(parsedNumbers[1]); // 1

// A stringified object
let userData = '{ "name": "John", "age": 35, "isAdmin": false, "friends": [0,1,2,3] }';
let user = JSON.parse(userData);
console.log(user.friends[1]); // 1
```

# Explanations

## Exercise 1:
`JSON.stringify(participant)` examines the object and translates all of its keys and values into a single contiguous text string formatted under strict JSON rules (like adding double-quotes around the keys `"id"`, `"name"`, etc).

Once it is a string, it can be passed over the internet or saved in text files. Later, when you retrieve it, `JSON.parse(jsonString)` reads that text string and reconstructs the real JavaScript object in memory so you can use it again in your code.
