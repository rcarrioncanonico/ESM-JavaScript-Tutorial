# Lesson 6 — Object Methods and "this" (JavaScript)

Objects are usually created to represent entities of the real world, like users, orders, or elements in a game. These objects can act and do things using functions associated with them, which we call *methods*.

## Learning objectives

By the end of this lesson you can:
- Add methods (functions) to objects.
- Use the `this` keyword to access the current object from inside a method.
- Understand method chaining.
- Avoid common mistakes when determining what `this` refers to.

## 1) Methods

A function that is a property of an object is called its **method**.

```javascript
let user = {
  name: "John",
  age: 30,
}; // Now we add a method to it

user.sayHi = function() {
  console.log("Hello!");
};

user.sayHi(); // Hello!

// We can also define it right inside the object using shorthand syntax:
let player = {
  score: 100,
  displayScore() { // same as displayScore: function() {...}
    console.log("Score is high!");
  }
};
player.displayScore();
```

## 2) The "this" keyword

It's common that a method needs to access the object's properties to do its job. For example, `player.displayScore()` might need to read `score`. To access the object, a method can use the `this` keyword.

The value of `this` is the object "before dot", the one used to call the method.

```javascript
let gameSession = {
  trialNumber: 5,
  showTrial() {
    // "this" refers to the "gameSession" object
    console.log("Current trial: " + this.trialNumber);
  },
  nextTrial() {
    this.trialNumber = this.trialNumber + 1;
  }
};

gameSession.showTrial(); // Current trial: 5
gameSession.nextTrial();
gameSession.showTrial(); // Current trial: 6
```

If we tried to use `gameSession.trialNumber` instead of `this.trialNumber` inside the method, it would work, but it is unreliable. If we later copied the object to another variable and deleted `gameSession`, the method would break! `this` ensures it always points to the *current* object executing the method.

## 3) Arrow functions have no "this"

Arrow functions are special: they don't have their "own" `this`. If we reference `this` from such a function, it's taken from the external "normal" function context.

```javascript
let userObj = {
  firstName: "Ilya",
  sayHi() {
    let arrow = () => console.log(this.firstName);
    arrow(); // Works because the arrow function uses the 'this' from the outer sayHi()
  }
};

userObj.sayHi(); // Ilya
```

# Solutions

## Exercise 1

```javascript
let calculator = {
  read(a, b) {
    this.a = a;
    this.b = b;
  },
  sum() {
    return this.a + this.b;
  },
  mul() {
    return this.a * this.b;
  }
};

calculator.read(5, 5);
console.log("Sum:", calculator.sum());
console.log("Product:", calculator.mul());
```

## Exercise 2

```javascript
let ladder = {
  step: 0,
  up() {
    this.step++;
    return this;
  },
  down() {
    this.step--;
    return this;
  },
  showStep() {
    console.log(this.step);
    return this;
  }
};

ladder.up().up().down().showStep(); // 1
```

# Explanations

## Exercise 1:
We use `this.a` and `this.b` to store the arguments inside the currently active object itself. Later, the `sum()` and `mul()` methods can read `this.a` and `this.b` to perform their operations. Without `this.a`, the method would look for a local variable `a` inside `sum()` and fail.

## Exercise 2:
The secret to **chaining** methods is returning the object itself (`return this;`) at the end of every method. When we write `ladder.up()`, it evaluates to `ladder`, meaning we can immediately call `.up()` on it again. This returns `ladder` once more, allowing us to seamlessly append `.down()` and `.showStep()`.
