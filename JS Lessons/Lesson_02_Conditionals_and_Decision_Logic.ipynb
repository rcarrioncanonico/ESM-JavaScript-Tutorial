# Lesson 02 - Conditionals and Decision Logic (JavaScript)

After variables and data types, the next step is making decisions in code.

In ESM cognitive tasks, conditionals control what the participant sees and what feedback they get.

## Learning objectives

By the end of this lesson, you can:

- Use `if`, `else if`, and `else` to control task behavior.
- Use comparison operators (`===`, `!==`, `>`, `<`, `>=`, `<=`).
- Use logical operators (`&&`, `||`) for multi-condition checks.
- Apply conditionals to simple trial feedback in an ESM-style task.

## 1) Core idea: `if / else`

Conditionals let your task choose what happens next.

- If a condition is true, run one block.
- Otherwise, run another block.

This is essential for correctness checks, trial flow, and feedback.

````typescript
// Notebook runtime example (Deno)
const expectedKey = "SPACE";
const pressedKey = "SPACE";

let feedback = "";

if (pressedKey === expectedKey) {
  feedback = "Correct";
} else {
  feedback = "Try again";
}

feedback
````

## 2) Comparison + logical operators

- `===` exactly equal (preferred)
- `!==` not equal
- `>` and `<` greater/less than
- `>=` and `<=` greater/less than or equal
- `&&` both conditions must be true
- `||` at least one condition must be true

Use `===` for safer comparisons in beginner code.

````typescript
// Mini practice: trial status message
const trialIndex = 3;
const maxTrials = 5;
const isPaused = false;

let status = "";

if (isPaused) {
  status = "Task paused";
} else if (trialIndex >= maxTrials) {
  status = "Task complete";
} else if (trialIndex > 0 && trialIndex < maxTrials) {
  status = "Task in progress";
} else {
  status = "Task not started";
}

status
````

## 3) Apply to your `student_task.js`

### Task objective
Show one of these messages based on trial state:

- `"Waiting to start"` when `trialIndex` is `0`
- `"Trial running"` when `trialIndex` is between `1` and `maxTrials - 1`
- `"Task complete"` when `trialIndex` reaches `maxTrials`

### Required code elements
- `if / else if / else`
- At least one comparison operator
- At least one logical operator (`&&` or `||`)
- A visible text label updated with `.setText(...)`

### Starter snippet to paste into `tasks/student_task.js`

```javascript
let trialIndex = 0;
const maxTrials = 5;

const statusLabel = this.add.text(400, 220, "", {
  fontSize: "28px",
  color: "#ffffff"
}).setOrigin(0.5);

function updateStatus() {
  if (trialIndex === 0) {
    statusLabel.setText("Waiting to start");
  } else if (trialIndex > 0 && trialIndex < maxTrials) {
    statusLabel.setText("Trial running");
  } else {
    statusLabel.setText("Task complete");
  }
}

updateStatus();
this.input.keyboard.on("keydown-SPACE", () => {
  if (trialIndex < maxTrials) trialIndex = trialIndex + 1;
  updateStatus();
});
```


````typescript
// Runner cell (do not edit)
const taskPath = "./tasks/student_task.js";
const taskCode = await Deno.readTextFile(taskPath);
play(taskCode);
````

## Success criteria

1. Message starts as `Waiting to start`.
2. Pressing `SPACE` shows `Trial running` before final trial.
3. At max trial, message becomes `Task complete`.
