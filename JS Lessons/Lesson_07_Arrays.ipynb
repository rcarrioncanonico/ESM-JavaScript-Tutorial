# Lesson 7 — Arrays (JavaScript)

Objects allow you to store keyed collections of values. But quite often we find that we need an *ordered collection*, where we have a 1st, a 2nd, a 3rd element and so on. For example, we need that to store a list of levels, timestamps, or trial configurations in a behavioral task. 

There exists a special data structure named `Array`, to store ordered collections.

## Learning objectives

By the end of this lesson you can:
- Declare arrays and access elements by their numeric index.
- Use the `length` property to count elements.
- Understand array methods acting as queues or stacks (`push`, `pop`, `shift`, `unshift`).
- Iterate through arrays with loops like `for...of`.

## 1) Declaration and Access

Arrays are mostly created with square brackets `[]`. Array elements are numbered, starting with zero. We can get an element by its number in square brackets.

````javascript
let fruits = ["Apple", "Orange", "Plum"];

console.log(fruits[0]); // Apple
console.log(fruits[1]); // Orange

// We can replace an element:
fruits[2] = "Pear"; // now ["Apple", "Orange", "Pear"]

// Or add a new one to the array:
fruits[3] = "Lemon"; // now ["Apple", "Orange", "Pear", "Lemon"]

// The total count of the elements in the array is its length:
console.log(fruits.length); // 4
````

## 2) Methods: push/pop, shift/unshift

Arrays allow acting as a queue or a stack. We can add or remove elements at the beginning or at the end.
- **push:** adds an item to the end.
- **pop:** takes an item from the end.
- **shift:** takes an item from the beginning.
- **unshift:** adds an item to the beginning.

````javascript
let queue = ["Apple", "Orange", "Pear"];

// pop: Extracts the last element of the array and returns it
let last = queue.pop(); 
console.log("Popped: " + last); // Pear

// push: Append the element to the end of the array
queue.push("Peach");

// shift: Extracts the first element of the array and returns it
let first = queue.shift();
console.log("Shifted: " + first); // Apple

// unshift: Add elements to the beginning of the array
queue.unshift("Pineapple", "Lemon");

console.log(queue); // ["Pineapple", "Lemon", "Orange", "Peach"]
````

## 3) Loops

One of the oldest ways to cycle array items is the `for` loop over indexes. But for arrays there is another form of loop, `for...of`, which is shorter and cleaner if we don't need the index.

````javascript
let arr = ["Apple", "Orange", "Pear"];

// The old-fashioned way (using index i)
for (let i = 0; i < arr.length; i++) {
  console.log("For loop: " + arr[i]);
}

console.log("-------");

// The modern, shorter way to iterate over array values
for (let fruit of arr) {
  console.log("For...of: " + fruit);
}
````

# Solutions

## Exercise 1

````javascript
let styles = ["Jazz", "Blues"];
styles.push("Rock-n-Roll");

// Replace middle element (works for any odd length)
let midIndex = Math.floor((styles.length - 1) / 2);
styles[midIndex] = "Classics";

let shifted = styles.shift();
console.log(shifted);

styles.unshift("Rap", "Reggae");
console.log(styles);
````

## Exercise 2

````javascript
function sumArray(arr) {
  let sum = 0;
  for (let num of arr) {
    sum += num;
  }
  return sum;
}

console.log(sumArray([1, 2, 3])); // 6
console.log(sumArray([-1, 10, 5, 2])); // 16
````

# Explanations

## Exercise 1:
Finding the middle index dynamically requires knowing the total length. `Math.floor((styles.length - 1) / 2)` safely extracts the exact integer index of the center item. The `push` method drops elements onto the end, while `shift` effectively slides the whole collection backward to remove the first element, and `unshift` slides them forward to make room at the front.

## Exercise 2:
A simple `for...of` loop is the cleanest approach when we don't care about the element indexes and just want to read values. We instantiate a `sum` variable initialized to 0, then continuously add the `num` variable—which iterates through each entry in the array sequentially.
