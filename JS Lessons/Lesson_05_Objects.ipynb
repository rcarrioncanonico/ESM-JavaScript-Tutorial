# Lesson 5 — Objects (JavaScript)

Objects are used to store keyed collections of various data and more complex entities. In JavaScript, objects penetrate almost every aspect of the language. To master JavaScript, you must understand objects.

## Learning objectives

By the end of this lesson you can:
- Create an object and define its properties.
- Access, modify, and delete object properties.
- Choose between dot notation (`obj.key`) and bracket notation (`obj["key"]`).
- Check if a property exists using the `in` operator.
- Iterate over an object's keys using the `for...in` loop.

## 1) Object Creation & Properties

An object can be created using curly braces `{...}`. A property is a "key: value" pair, where the key is a string (also called a "property name"), and the value can be anything.

````javascript
let user = {     // an object
  name: "John",  // by key "name" store value "John"
  age: 30        // by key "age" store value 30
};

// Accessing properties via dot notation:
console.log(user.name);

// Adding a new property:
user.isAdmin = true;

// Deleting a property:
delete user.age;

console.log(user);
````

## 2) Square Brackets

For multi-word properties or dynamically calculated keys, dot notation doesn't work. We must use square brackets and quote the key.

````javascript
let task = {
  "stimulus duration": 500
};

// task.stimulus duration = 500; // Syntax error!
task["stimulus duration"] = 1000; // Works!

// Bracket notation is also useful when the key is stored in a variable:
let key = "isAdmin";
let player = {};
player[key] = true;

console.log(player.isAdmin); // true
````

## 3) The `in` Operator and `for...in` Loop

You can check if a property exists using the `in` operator.
To walk over all keys of an object, there exists a special form of the loop: `for...in`.

````javascript
let settings = {
  volume: 80,
  brightness: 50
};

console.log("volume" in settings); // true
console.log("difficulty" in settings); // false

for (let key in settings) {
  // "key" takes the property names (volume, brightness)
  console.log(key + ": " + settings[key]);
}
````

## Exercise 2
